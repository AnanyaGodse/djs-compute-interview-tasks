```python
import pandas as pd
from datetime import datetime
from lazypredict.Supervised import LazyRegressor
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('C:/Users/Ananya/Desktop/breast cancer/breast_cancer_survival_cleaned.csv')
data
```

```python
data[['Date_of_Surgery','Date_of_Last_Visit']] = data[['Date_of_Surgery','Date_of_Last_Visit']].apply(pd.to_datetime)
```

```python
X = data.drop(columns=['Patient_Status'])
X
```

```python
y = data['Patient_Status']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
reg_model = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg_model.fit(X_train, X_test, y_train, y_test)
```
