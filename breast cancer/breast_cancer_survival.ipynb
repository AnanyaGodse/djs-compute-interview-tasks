```python
df = df[df['Patient_Status'].notnull()]
df
```

```python
df.shape
```

```python
df = df.rename(columns={'ER status':'ER_Status', 'PR status':'PR_Status', 'HER2 status':'HER2_Status', 'Surgery_type':'Surgery_Type' })
df
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].replace(['FEMALE','MALE'],['1','0'])
df
```

```python
df['Tumour_Stage'].value_counts()
```

```python
df['Tumour_Stage'] = df['Tumour_Stage'].replace(['I','II','III'],['1','2','3'])
df
```

```python
df['Histology'].value_counts()
```

```python
df['Histology'] = df['Histology'].replace(['Infiltrating Ductal Carcinoma','Infiltrating Lobular Carcinoma','Mucinous Carcinoma'],['1','2','3'])
df
```

```python
df['ER_Status'].value_counts()
```

```python
df['ER_Status'] = df['ER_Status'].replace(['Positive'],'1')
df
```

```python
df['PR_Status'].value_counts()
```

```python
df['PR_Status'] = df['PR_Status'].replace(['Positive'],'1')
df
```

```python
df['HER2_Status'].value_counts()
```

```python
df['HER2_Status'] = df['HER2_Status'].replace(['Positive','Negative'],['1','2'])
df
```

```python
df['Surgery_Type'].value_counts()
```

```python
df['Surgery_Type'] = df['Surgery_Type'].replace(['Simple Mastectomy','Lumpectomy','Modified Radical Mastectomy','Other'],['1','2','3','4'])
df
```

```python
df['Patient_Status'].value_counts()
```

```python
df['Patient_Status'] = df['Patient_Status'].replace(['Alive','Dead'],['1','0'])
df
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
df[['Date_of_Surgery','Date_of_Last_Visit']] = df[['Date_of_Surgery','Date_of_Last_Visit']].apply(pd.to_datetime)
df
```

```python
df.dtypes
```

```python
df = df.sort_values(by='Age')
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.1f')
plt.show()
```
